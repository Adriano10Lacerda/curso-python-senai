```python
# Criar e manipular arquivos em excel
import xlsxwriter as xlsx
# Para gerenciar arquivos no sitema operacional
import os 

#-------------------------Gerar um caminho e um arquivo 

caminho = 'C:\\Users\\FIC\\Documents\\Adriano-Python\\'
arquivo = 'Exemplo 05.xlsx'

workbook = xlsx.Workbook (caminho + arquivo)
sheetDados = workbook.add_worksheet('Relatorio')
#--------------------------Modificando a cor de fundo das celulas


#--------------------------Listar dados

dados = [
    ['Produto', 'Qtde', 'Preço', 'Subtotal' ],
    ['calça', 5, 50],
    ['Camisa', 4, 30],
    ['Sapato', 5, 40],
    ['Saia', 8, 20],
    ['Blusa', 6, 10],
    ['Shorts', 8, 25]
]
#--------------------------Definir regras de formatação
#--------------------------Formatação titulos
titles = workbook.add_format({
                            'bg_color':'blue',
                            'font_color':'white',
                            'font_name':'Arial',
                            'align':'center',
                            'bold':True
                            })

rotulosLinhas = workbook.add_format({
                                    'font_color':'gray',
                                    'font_name':'Arial',
                                    'align':'left'                                                                      
                                    })

qtde = workbook.add_format({
                           'font_color':'gray',
                           'font_name':'Arial',
                           'align':'center',
                           'num_format':'R$*#.##0,0'
                           })

valores = workbook.add_format({
                           'font_color':'gray',
                           'font_name':'Arial',
                           'align':'center',
                           'num_format':'R$*#.##0,0'
                           })

formatoDatas = workbook.add_format({
                                 'font_color':'gray',
                                 'font_name':'Arial',
                                 'align':'center',
                                 'num_format': 'dd/mm/yyyy'
                                  })

formatoHora = workbook.add_format({
                                 'font_color':'gray',
                                 'font_name':'Arial',
                                 'align':'center',
                                 'num_format': 'h:m'
                                  })

mesclado = workbook.add_format({
                                 'font_color':'gray',
                                 'font_name':'Arial',
                                 'align':'center',
                                 'bold':True
                                  })

#--------------------------Registrar data
sheetDados.merge_range('A1:D1','Planilha de Solicitações', mesclado)

sheetDados.write('A3', 'Data')
sheetDados.write_formula('B3','=Today()', formatoDatas)
sheetDados.write('A4','Hora')
sheetDados.write_formula('B4', '=Now()', formatoHora)

#--------------------------Registrar os dados na planilha
posLinha = 5
posColuna = 0

for linha, listas in enumerate(dados):
    if linha == 0:
        sheetDados.write_row(linha + posLinha, posColuna, listas, titles)# Cabeçalho
    else:        
        sheetDados.write(linha + posLinha, posColuna, listas[0])# Produto
        sheetDados.write(linha + posLinha, posColuna, listas[1])# Qtde
        sheetDados.write(linha + posLinha, posColuna, listas[2], valores)# Preço Unitario
        sheetDados.write_formula(linha + posLinha, posColuna + 3, '=B' + str(linha + posLinha+1) + '*C'+ str(linha + posLinha+1))

#---------------------------Gerar Total
sheetDados.write('C13','Total')
sheetDados.write('D13', '=SUM(D7:D12)', valores)

#--------------------------Fechar e salvar
workbook.close()
#--------------------------Abrir o documento
os.startfile(caminho + arquivo)   
```

```python
#-------------------------IMPORTAR AS BIBLIOTECAS
import xlsxwriter as xlsx
import os

#--------------------------Caminho do arquivo

caminho = 'C:\\Users\\FIC\\Documents\\Adriano aula 2 python\\'
arquivo = 'Exemplo 03.xlsx' # Cria uma nova pasta de trabalho

#--------------------------Adicionar uma planilha (aba) ao documento

workbook = xlsx.Workbook(caminho + arquivo)
sheetDados = workbook.add_worksheet('Dados')

#--------------------------Modificando a cor de fundo das celulas

higherFormat = workbook.add_format({
                                    'bg_color':'#E2EFDA',
                                    'font_color':'green',
                                    'bold': True
                                  })
lowerFormat = workbook.add_format({
                                    'bg_color':'#F8CBAD',
                                    'font_color':'red',
                                    'bold':True
                                })

#-----------------------------Criar um modelo de dados para aplicar a formatação

dados = [
    ['Coluna1', 'Coluna2', 'Coluna3', 'Coluna4'],
    [10, 20, 30, 40],
    [41, 28, 56, 89],
    [78, 52, 12, 7],
    [1, 59, 71, 34]
]

for linha, listas in enumerate(dados):
    sheetDados.write_row(linha+1, 1,listas)
    
#-------------------------------Formatação condicional

sheetDados.conditional_format('A1:D5',{'type':'cell',
                                      'criteria':'>=',
                                      'value': 50,
                                      'format':higherFormat                                      
                                      }
                             )
sheetDados.conditional_format('A1:D5',{'type':'cell',
                                       'criteria':'<',
                                       'value': 50,
                                       'format':lowerFormat                                    
                                       }
                             )

#-------------------------------Fechar o arquivo e salvar
workbook.close()

#-------------------------------Abrir o documento

os.startfile(caminho + arquivo)
```
